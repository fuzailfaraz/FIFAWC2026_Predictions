# 01. Data Exploration
This notebook inspects and visualizes the raw datasets for the FIFA World Cup 2026 Analytics Platform.

```
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
train_df = pd.read_csv('../data/wc_training_dataset_updated.csv')
master_df = pd.read_csv('../data/wc2026_master_dataset.csv')
historical_df = pd.read_csv('../data/team_historical_features_updated.csv')

print(f"Training Dataset: {train_df.shape}")
print(f"Master 2026 Dataset: {master_df.shape}")
print(f"Historical Team Features: {historical_df.shape}")
```

### Class Distribution in Training Data
Check the distribution of match outcomes (H = Home Win, A = Away Win, D = Draw).

```
result_counts = train_df['result'].value_counts()
result_pct = train_df['result'].value_counts(normalize=True) * 100
for idx in result_counts.index:
    print(f"{idx}: {result_counts[idx]} matches ({result_pct[idx]:.1f}%)")

plt.figure(figsize=(6, 4))
plt.bar(result_counts.index, result_counts.values, color=['#3b82f6', '#94a3b8', '#f97316'])
plt.title('Distribution of World Cup Match Outcomes (1994-2018)')
plt.xlabel('Result')
plt.ylabel('Match Count')
plt.show()
```
